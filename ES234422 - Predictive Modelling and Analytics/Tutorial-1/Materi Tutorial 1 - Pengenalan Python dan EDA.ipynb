## **3. Praproses Data**
Data Preprocessing adalah kegiatan pengolahan data agar dapat dioperasikan secara lebih lanjut. Secara umum, kegiatan ini bisa disebut juga dengan "data cleansing" dimana kita mengolah data yang kotor (value yang hilang, duplikat, outlier) menjadi data bersih.


### Data kotor? Kotor bagaimana maksudnya?

- Data duplikat: satu baris data yang muncul lebih dari satu kali
- Data tidak lengkap: beberapa field berisi null value
- Data tidak konsisten: data dalam format yang salah atau format yang berbeda-beda

Jalankan kode berikut ini untuk membuat dummy data yang belum bersih.

```python
# Membuat data frame
nama = ["Aida", "Beta", "Charlie", "Doni", "Eka", "Firda", "Gina", "Hera", "Indra", "Janet", "Eka"]
nilai = [75, 66, 80, 33, 80, 99, 67, 56, 89, 75, 80]
kelas = [np.nan, np.nan, np.nan, np.nan, 'B', 'A', np.nan, np.nan, 'C', np.nan, 'B']
jurusan = ['IPA', 'IPA', 'IPS', 'IPA', 'IPS', np.nan, 'IPA', 'IPA', 'IPA', 'IPA', 'IPS']
ekonomi = ["Mampu", "Kurang Mampu", "Kurang Mampu", "Menengah", "Menengah", "Mampu", "Kurang Mampu", "Mampu", "Mampu", "Menengah", "Menengah"]
beasiswa = ["Tidak", "Ya", "Ya", "Tidak", "Tidak", "Ya", "Tidak", "Ya", "Tidak", "Ya", "Tidak"]

data_kelas = pd.DataFrame(data = {'nama':nama,
                                  'nilai':nilai,
                                  'kelas':kelas,
                                  'jurusan':jurusan,
                                  'ekonomi':ekonomi,
                                  'beasiswa':beasiswa})
data_kelas
```

### Pemeriksaan Missing Value

```python
# Cek apakah terdapat baris yang kosong
data_kelas.isnull().sum()
```

`isnull()` akan memberikan output berupa baris-baris data mana saja yang mengandung null value di dalamnya.

#### Mengatasi Missing Value

Untuk mengatasi missing value, beberapa solusi dapat dilakukan.
- Menghapus missing value/sparse columns (kolom yang berisi banyak null)
- Mengisi missing value dengan nilai mayoritas (mean, median, mode)

Berdasarkan informasi yang sebelumnya didapat, dalam kolom kelas hanya terdapat 4 baris data yang memiliki value. Hampir seluruhnya berisi null value. Maka, kita dapat menghapus kolom kelas.

```python
# Hapus kolom yang memiliki banyak nilai kosong
dk = data_kelas.drop(columns='kelas')
dk
```

Pada kolom jurusan, hanya terdapat 1 null value, tidak selayaknya apabila kita hapus keseluruhan kolom. Maka, kita dapat mengisi null value tersebut dengan nilai yang paling sering muncul.

```python
dk['jurusan'].value_counts()
```

`value_counts()` akan memberikan output berupa frekuensi suatu nilai muncul.

```python
# memberikan nilai pada baris kosong
jurusan_max = dk['jurusan'].value_counts().idxmax()
dk['jurusan'] = dk['jurusan'].fillna(jurusan_max)
dk
```

`idxmax()` memberikan output berupa index dari data tertinggi, dalam kasus ini adalah 'IPA'.

`fillna()` akan mengisi null value dengan nilai yang dipilih.

```python
# Periksa kembali data
dk.isnull().sum()
```

In missing-data imputation, the <u>*tradeoff*</u> between **dropping sparse columns** and **filling missing values**:

* *dropping sparse column* (column with lots of missing values): in some cases, it could be that this column may still hold essential information

* *filling missing values* with mode/mean/median: may not really represent the actual data

### Outlier

Outlier adalah data berjenis data interval, yang nilainya berbeda secara signifikan dari keseluruhan data. Outlier memiliki dampak besar pada hasil analisis statistik atau pemodelan data. Untuk mendeteksi outlier, kita dapat menggunakan box plot.

```python
# Box Plot
plt.figure(figsize = (5, 5))
dk_encoded['nilai'].plot(kind="box")
plt.show()
```

### Menghapus Outlier

It is suggested to remove outlier as these **excessively high/low values may not represent actual data** hence introducing **bias** to the model.

Other reference:
https://medium.com/analytics-vidhya/how-to-remove-outliers-for-machine-learning-24620c4657e8

#### 1. Menggunakan z score

image.png

Z-score berfungsi untuk mengetahui apakah suatu nilai lebih rendah dari nilai rataan dalam satu dataset dan mengukur seberapa jauh dari rataan

```python
dk_mean = dk_encoded['nilai'].mean()
dk_std = dk_encoded['nilai'].std()
print("Nilai rata-rata: {0}\nNilai Standard Deviasi: {1}".format(dk_mean, dk_std))
```

```python
# threshold value for z-score
threshold = 1
```

Threshold value 1 berarti kita hanya menyimpan nilai yang berjarak dalam 1 standar deviasi dari rataan.

```python
outlier = []

for i in dk_encoded['nilai']:
    z = (i-dk_mean)/dk_std
    if abs(z) > threshold:
        outlier.append(i)

print('Outlier dari kolom nilai adalah', outlier)
```

Z-score juga dapat dihitung dengan library Scipy, dengan `stats.zscore()`

```python
from scipy import stats
dk_encoded['z_score'] = stats.zscore(dk_encoded['nilai'])
dk_encoded.head()
```

```python
dk_no_outliers = dk_encoded.loc[dk_encoded['z_score'].abs() < threshold]
dk_no_outliers = dk_no_outliers.reset_index(drop=True)
dk_no_outliers
```

#### 2. Inter Quartile Range (IQR)

image.png

IQR (interquartile range) adalah jarak antara kuartil bawah dan kuartil atas. Dalam metode IQR, kita dapat menemukan outlier dengan menghitung batas bawah dan batas atas. Data diluar batas atas dan bawah akan dianggap sebagai outlier.

```python
def outlier_treatment(datacolumn):
  sorted(datacolumn)  # Memastikan data sudah urut
  Q1, Q3 = np.percentile(datacolumn , [25,75])
  IQR = Q3 - Q1
  lower_range = Q1 - (1.5 * IQR)
  upper_range = Q3 + (1.5 * IQR)
  return lower_range, upper_range
```

```python
lowerbound, upperbound = outlier_treatment(dk_encoded.nilai)
print("Batas bawah: {0}\nBatas atas: {1}".format(lowerbound, upperbound))
```

```python
# Cek data yang menjadi outlier
dk_encoded[(dk_encoded.nilai < lowerbound) | (dk_encoded.nilai > upperbound)]
```

```python
# Menghapus outlier
dk_no_outlier = dk_encoded.drop(dk_encoded[(dk_encoded.nilai < lowerbound) | (dk_encoded.nilai > upperbound)].index)
dk_no_outlier
```

### Removing Stopwords

Salah satu langkah penting ketika memproses data teks adalah menghapus stopwords. Stopwords adalah kata-kata yang paling umum dalam bahasa dan jarang memberikan informasi yang berguna untuk proses klasifikasi.

```python
def remove_stopwords(sentence):
    # List of stopwords
    stopwords = ["a", "about", "above", "after", "again", "against", "all", "am", "an", "and", "any", "are", "as", "at", "be", "because", "been", "before", "being", "below", "between", "both", "but", "by", "could", "did", "do", "does", "doing", "down", "during", "each", "few", "for", "from", "further", "had", "has", "have", "having", "he", "he'd", "he'll", "he's", "her", "here", "here's", "hers", "herself", "him", "himself", "his", "how", "how's", "i", "i'd", "i'll", "i'm", "i've", "if", "in", "into", "is", "it", "it's", "its", "itself", "let's", "me", "more", "most", "my", "myself", "nor", "of", "on", "once", "only", "or", "other", "ought", "our", "ours", "ourselves", "out", "over", "own", "same", "she", "she'd", "she'll", "she's", "should", "so", "some", "such", "than", "that", "that's", "the", "their", "theirs", "them", "themselves", "then", "there", "there's", "these", "they", "they'd", "they'll", "they're", "they've", "this", "those", "through", "to", "too", "under", "until", "up", "very", "was", "we", "we'd", "we'll", "we're", "we've", "were", "what", "what's", "when", "when's", "where", "where's", "which", "while", "who", "who's", "whom", "why", "why's", "with", "would", "you", "you'd", "you'll", "you're", "you've", "your", "yours", "yourself", "yourselves" ]

    # Sentence converted to lowercase-only
    sentence = sentence.lower()

    words = sentence.split()
    cleaned_words = [word for word in words if word not in stopwords]
    sentence = ' '.join(cleaned_words)

    return sentence
```

```python
remove_stopwords("I am about to go to the store and get any snack")
```

### Reading the raw data

Sekarang kita perlu membaca data dari file csv.

```python
def parse_data_from_file(filename):
    sentences = []
    labels = []

    with open(filename, 'r') as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        next(reader, None)

        for row in reader:
            sentences.append(remove_stopwords(row[1]))
            labels.append(remove_stopwords(row[0]))

    return sentences, labels
```

```python
# With original dataset
sentences, labels = parse_data_from_file("bbc-text.csv")

print(f"There are {len(sentences)} sentences in the dataset.\n")
print(f"First sentence has {len(sentences[0].split())} words (after removing stopwords).\n")
print(f"There are {len(labels)} labels in the dataset.\n")
print(f"The first 5 labels are {labels[:5]}\n\n")
```

### Using the Tokenizer

Proses selanjutnya adalah tokenisasi. Tokenisasi digunakan dalam pemrosesan bahasa alami untuk membagi paragraf dan kalimat menjadi unit-unit yang lebih kecil sehingga lebih mudah diberi makna. Langkah pertama dari proses NLP adalah mengumpulkan data (kalimat) dan memecahnya menjadi bagian-bagian (kata-kata) yang dapat dimengerti.

![Tokenization](https://miro.medium.com/v2/resize:fit:1200/1*UhfwmhMN9sdfcWIbO5_tGg.jpeg)

```python
def fit_tokenizer(sentences):
    # Instantiate the Tokenizer class by passing in the oov_token argument
    tokenizer = Tokenizer(oov_token='<OOV>')
    # Fit on the sentences
    tokenizer.fit_on_texts(sentences)
    return tokenizer
```

```python
tokenizer = fit_tokenizer(sentences)
word_index = tokenizer.word_index

print(f"Vocabulary contains {len(word_index)} words\n")
print("<OOV> token included in vocabulary" if "<OOV>" in word_index else "<OOV> token NOT included in vocabulary")
```

Dalam melakukan pemodelan, model harus mendapat input yang memiliki bentuk dan ukuran yang sama. Namun, saat kita melakukan pra-pemrosesan dan menggunakan teks sebagai masukan untuk model kita, tidak semua kalimat memiliki panjang yang sama. Beberapa kalimat ada yang lebih panjang atau lebih pendek. Kita perlu memiliki input dengan ukuran yang sama, maka perlu dilakukan padding/truncating. Kita akan mencoba melakukan padding dengan kode dibawah ini.

```python
def get_padded_sequences(tokenizer, sentences):
    # Convert sentences to sequences
    sequences = tokenizer.texts_to_sequences(sentences)

    # Pad the sequences using the post padding strategy
    padded_sequences = pad_sequences(sequences, padding='post')

    return padded_sequences
```

```python
padded_sequences = get_padded_sequences(tokenizer, sentences)
print(f"First padded sequence looks like this: \n\n{padded_sequences[0]}\n")
print(f"Numpy array of all sequences has shape: {padded_sequences.shape}\n")
print(f"This means there are {padded_sequences.shape[0]} sequences in total and each one has a size of {padded_sequences.shape[1]}")
```

Label juga perlu ditokenisasi.

```python
def tokenize_labels(labels):
    # Instantiate the Tokenizer class
    # No need to pass additional arguments since you will be tokenizing the labels
    label_tokenizer = Tokenizer()

    # Fit the tokenizer to the labels
    label_tokenizer.fit_on_texts(labels)

    # Save the word index
    label_word_index = label_tokenizer.word_index

    # Save the sequences
    label_sequences = label_tokenizer.texts_to_sequences(labels)

    return label_sequences, label_word_index
```

```python
label_sequences, label_word_index = tokenize_labels(labels)
print(f"Vocabulary of labels looks like this {label_word_index}\n")
print(f"First ten sequences {label_sequences[:10]}\n")
```

Setelah tokenisasi dan padding, maka data kita siap untuk digunakan dalam proses pemodelan!
